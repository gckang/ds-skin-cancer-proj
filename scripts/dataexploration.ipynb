```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

#creating data frame
df = pd.read_csv('metadata.csv')
# isolating skin cancer counts
    diagnostic_counts = df['diagnostic'].value_counts()
# plotting graph of types of skin cancer
plt.figure(figsize=(12, 6))
colors = list(mcolors.TABLEAU_COLORS.values())
sns.barplot(x=diagnostic_counts.index, y=diagnostic_counts.values, palette=colors[:len(diagnostic_counts)])
plt.xlabel("Skin Cancer")
plt.ylabel("Count")
plt.title("Skin Cancer Types")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()
plt.show()
```

```python
#gender graph
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df, palette=['lightblue', 'lightcoral']) # Set colors for male and female
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender")
plt.show()
```

```python
# skin leison graph
plt.figure(figsize=(10, 6))
sns.countplot(x='region', data=df, palette='viridis')
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Distribution of Regions")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
